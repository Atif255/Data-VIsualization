```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("CardioGoodFitness-1.csv")
```

```python
df
```

## Relplot

```python
sns.relplot(x="Age",y="Income",data=df)
```

#It can be inferred that upto a certain age (say 30 or 35), income and age can be correlated.The readings are clustered in under that region.

```python
sns.relplot(x="Age",y="Income",hue="Gender",data=df)
```

```python
sns.relplot(x="Education",y="Income",hue="Gender",data=df)
```

```python
sns.relplot(x="Education",y="Income",hue="Age",data=df)
```

```python
sns.relplot(x="Age",y="Miles",hue="Gender",data=df)
```

```python
sns.relplot(x="Age",y="Miles",hue="Fitness",data=df)
```

```python
sns.relplot(x="Age",y="Miles",hue="Fitness",size="Income",data=df)
```

## Catplot

```python
sns.catplot(x="Gender",y="Income",data=df)
```

```python
sns.catplot(x="Gender",y="Education",data=df)
```

```python
sns.catplot(x="Gender",y="Income",kind="swarm",data=df)
```

```python
sns.catplot(x='Product', y='Age', data=df)
```

```python
sns.catplot(x='Product', y='Age',jitter=False, data=df)
```

```python
sns.catplot(x='Product', y='Age',hue='Education', data=df)
```

```python
sns.catplot(x='Product', y='Age',hue='MaritalStatus',kind='swarm', data=df)
```

```python
sns.catplot(x="Gender",y="Income",kind="box",data=df)
```

```python
sns.catplot(x='Education', y='Income',kind='box', data=df)
```

```python
sns.catplot(x="Education",y="Income",kind="box",hue="MaritalStatus",data=df)
```

```python
sns.catplot(x="Gender",y="Income",kind="box",hue="Education",data=df)
```

```python
sns.catplot(x="Gender",y="Income",kind="violin",data=df)
```

```python
sns.catplot(x='MaritalStatus',y='Education',kind='bar',hue='Gender',data=df)
```

```python
sns.catplot(x='MaritalStatus',y='Income',kind='swarm',hue='Gender',col='Product',data=df)
```
