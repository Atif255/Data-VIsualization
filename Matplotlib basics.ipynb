### Legends

```python
days= [1,2,3,4,5,6,7]
max_t= [50,51,53,49,52,47,46]
min_t= [43,42,40,41,33,33,36]
avg_t= [45,46,48,44,42,41,39]
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t,label="Avg")
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.legend(loc="best",shadow="True",fontsize="large")
plt.grid()
```

# BAR CHART

```python
company= ["google","amazon","microsoft","facebook"]
revenue=[89,135,91,26]
profit=[40,3,35,13]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="Profit")
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.2, revenue,height=0.4, label="Revenue")
plt.barh(xpos+0.2, profit,height=0.4, label="Profit")
plt.legend()
```

# Histogram

```python
blood_sugar_men=[149,151,85,91,114,89,94,116,136,81,76,83,130]
blood_sugar_women = [90,99,65,124,136,150,70,90,85,110,115,120]
plt.xlabel("sugar level")
plt.ylabel("no of patients")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[60,100,125,200],rwidth=0.95,color=["green","orange"],label=["men","women"])
plt.legend()
```

```python
blood_sugar_men=[149,151,85,91,114,89,94,116,136,81,76,83,130]
blood_sugar_women = [90,99,65,124,136,150,70,90,85,110,115,120]
plt.xlabel("sugar level")
plt.ylabel("no of patients")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[60,100,125,200],rwidth=0.95,color=["green","orange"],label=["men","women"],orientation="horizontal")
plt.legend()
```

# Pie Chart

```python
expensevalue=[1500,400,800,500,300]
label=["rent","food","internet bill","transport","others"]
```

```python
plt.axis("equal")
plt.pie(expensevalue,labels=label,radius=1.75,autopct="%0.0f%%",explode=[0,0,0,1,0])
plt.show()
```
